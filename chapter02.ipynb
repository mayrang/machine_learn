```python
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from zlib import crc32
from sklearn.model_selection import train_test_split
from pandas.plotting import scatter_matrix
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import rbf_kernel
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import FunctionTransformer, StandardScaler
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.compose import ColumnTransformer, make_column_selector
from sklearn.metrics import mean_squared_error
from sklearn.cluster import KMeans
from sklearn.utils.validation import check_array, check_is_fitted
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV


def load_housing_data():

  return pd.read_csv(Path("datasets/housing/housing.csv"))


housing = load_housing_data()
print("1. Data acquisiion")
print(housing)

print("2. a quick look")
print("head() method")
print(housing.head())
print("info() method")
print(housing.info())
print("value_counts() method")
print(housing["ocean_proximity"].value_counts())
print("describe() method")
print(housing.describe())

housing.hist(bins=50, figsize=(15, 10))
plt.show()

def shuffle_and_split_data_rand_fixed(data, test_ratio, seed_seq):
    np.random.seed(seed_seq)
    num_samples = len(data)
    indices = np.random.permutation(num_samples)
    num_test_samples = int(test_ratio * num_samples)
    train_indices = indices[num_test_samples:]
    test_indices = indices[:num_test_samples]
    return data.iloc[train_indices], data.iloc[test_indices]

print("3. set separation")
train_set_2, test_set_2 = shuffle_and_split_data_rand_fixed(housing, 0.2,  42)
print("train_set_2")
print(train_set_2)
print("test_set_2")
print(test_set_2)


print("4. Training/test set separation")

def is_id_in_test_set(id, test_ratio):
   flag_test = crc32(np.int64(id)) < test_ratio * 2**32
   return flag_test

def split_data_with_id_hash(data, test_ratio, name_col_ids):
   ids = data[name_col_ids]
   flag_test_all = np.zeros(len(data), dtype=bool)
   for idx, val in enumerate(ids):
     flag_test_all[idx] = is_id_in_test_set(val, test_ratio)

   return data.loc[~flag_test_all], data.loc[flag_test_all]

housing_with_default = housing.reset_index().copy()


train_set_3, test_set_3 = split_data_with_id_hash(housing_with_default, 0.2, "index")
print("train_set_3")
print(train_set_3)
print("test_set_3")
print(test_set_3)

housing["income_cat"] = pd.cut(housing["median_income"], bins=[0., 1.5, 3.0, 4.5, 6., np.inf], labels=[1, 2, 3, 4, 5])


train_set_6, test_set_6 = train_test_split(housing, test_size=0.2, stratify=housing["income_cat"], random_state=42)

print("train_set_6")
print(train_set_6)
print("test_set_6")
print(test_set_6)


housing.plot(kind="scatter", x="longitude", y="latitude", grid=True, s=housing["population"] / 100, label="population", c="median_house_value", cmap="jet", colorbar=True,  legend=True, sharex=False, figsize=(10, 7))
plt.show()



corr_mat = housing.corr(numeric_only=True)
print("corr_median_house_value_sorted:")
print(corr_mat["median_house_value"].sort_values(ascending=False))

attributes = ["median_house_value", "median_income", "total_rooms", "housing_median_age"]
scatter_matrix(housing[attributes], figsize=(12, 8))
plt.show()


train_set_new_labels = train_set_6["median_house_value"].copy(deep=True)
train_set_new = train_set_6.drop(["median_house_value"], axis=1)
test_set_new_labels = test_set_6["median_house_value"].copy(deep=True)
test_set_new = test_set_6.drop(["median_house_value"], axis=1)

housing["rooms_per_house"] = housing["total_rooms"] / housing["households"]
housing["bedrooms_ratio"] = housing["total_bedrooms"] / housing["total_rooms"]
housing["rooms_per_house"] = housing["population"] / housing["households"]

print("corr_median_house_value_sorted:")
corr_mat = housing.corr(numeric_only=True)
print(corr_mat["median_house_value"].sort_values(ascending=False))


dropna_train_set_new = train_set_new.dropna()
print(dropna_train_set_new.info())
dropna_train_set_new_feature = train_set_new.dropna(axis=1)
print(dropna_train_set_new_feature.info())

imputer = SimpleImputer(strategy="median")
train_set_6_new_num = train_set_new.select_dtypes(include=[np.number])
imputer.fit(train_set_6_new_num)
print("imputer.statistics_:")
print(imputer.statistics_)
print("train_set_6_new_num.median().values:")
print(train_set_6_new_num.median().values)
print("train_set_6_new_num_imputed")
train_set_6_new_num_imputed = train_set_6_new_num.info()
print(train_set_6_new_num_imputed)


print("train_set_6_new_cat.head(8) : ")
train_set_6_new_cat = train_set_new[["ocean_proximity"]]
print(train_set_6_new_cat.head(8))
print("train_set_6_new_cat_ordinal:")
ordinal_encoder = OrdinalEncoder()
train_set_6_new_cat_ordinal = ordinal_encoder.fit_transform(train_set_6_new_cat)
print(train_set_6_new_cat_ordinal[:8])

print("train_set_6_new_cat_onehot:")
cat_encoder = OneHotEncoder()
train_set_6_new_cat_onehot = cat_encoder.fit_transform(train_set_6_new_cat)
print(train_set_6_new_cat_onehot[:8])
print(train_set_6_new_cat_onehot[:8].toarray())

print(train_set_6_new_cat)
print("train_set_6_new_num_imputed_min_max_scaled:")
min_max_scaler = MinMaxScaler(feature_range=(-1, 1))
train_set_6_new_num_imputed_min_max_scaled = min_max_scaler.fit_transform(train_set_6_new_num)
print(train_set_6_new_num_imputed_min_max_scaled)

print("train_set_6_new_num_imputed_std_scaled:")
std_scaler = StandardScaler()
train_set_6_new_num_imputed_std_scaled = std_scaler.fit_transform(train_set_6_new_num)
print(train_set_6_new_num_imputed_std_scaled)


train_set_6['population_log'] = np.log(train_set_6['population'])
plt.subplot(1, 2, 1)
train_set_6['population'].hist(bins=50)
plt.xlabel('Population')
plt.ylim(0, max(np.histogram(train_set_6['population'], bins=50)[0].max(), np.histogram(train_set_6['population_log'], bins=50)[0].max()))


plt.subplot(1, 2, 2)
train_set_6['population_log'].hist(bins=50)
plt.xlabel('Log of Population')
plt.ylim(0, max(np.histogram(train_set_6['population'], bins=50)[0].max(), np.histogram(train_set_6['population_log'], bins=50)[0].max()))
plt.ylabel('Number of districts')


plt.tight_layout()
plt.show()



train_set_6['population_sqrt'] = np.sqrt(train_set_6['population'])
plt.subplot(1, 2, 1)
train_set_6['population'].hist(bins=50)
plt.xlabel('Population')
plt.ylim(0, max(np.histogram(train_set_6['population'], bins=50)[0].max(), np.histogram(train_set_6['population_sqrt'], bins=50)[0].max()))


plt.subplot(1, 2, 2)
train_set_6['population_sqrt'].hist(bins=50)
plt.xlabel('sqrt(Population)')
plt.ylim(0, max(np.histogram(train_set_6['population'], bins=50)[0].max(), np.histogram(train_set_6['population_sqrt'], bins=50)[0].max()))
plt.ylabel('Number of districts')


plt.tight_layout()
plt.show()



percentile_1 = np.percentile(train_set_6['median_income'], np.arange(0, 101, 1))
train_set_6['median_income_bucket'] = pd.cut(train_set_6['median_income'], bins=percentile_1, labels=False, include_lowest=True)

plt.subplot(1, 2, 1)
train_set_6['median_income'].hist(bins=50)
plt.xlabel('Median income')
plt.ylim(0, 2000)


plt.subplot(1, 2, 2)
train_set_6['median_income_bucket'].hist(bins=100)
plt.xlabel('Median income percentile')
plt.ylim(0, 2000)
plt.ylabel('Number of districts')


plt.tight_layout()
plt.show()

percentile_20 = np.percentile(train_set_6['housing_median_age'], np.arange(0, 101, 20))
train_set_6['housing_median_age_bucket'] = pd.cut(train_set_6['housing_median_age'], bins=percentile_20, labels=False, include_lowest=True)


plt.subplot(1, 2, 1)
train_set_6['housing_median_age'].hist(bins=50)
plt.xlabel('Median age')



plt.subplot(1, 2, 2)
train_set_6['housing_median_age_bucket'].hist(bins=5)
plt.xlabel('Median income percentile')
plt.ylabel('Number of districts')


plt.tight_layout()
plt.show()


ages = np.linspace(train_set_6["housing_median_age"].min(),
                   train_set_6["housing_median_age"].max(),
                    len(train_set_6)).reshape(-1, 1)
gamma1 = 0.1

rbf1 = rbf_kernel(ages, [[16]], gamma=gamma1)
rbf2 = rbf_kernel(ages, [[26]], gamma=gamma1)
rbf3 = rbf_kernel(ages, [[35]], gamma=gamma1)


train_set_6['age_simil_16'] = rbf1.flatten()
train_set_6['age_simil_26'] = rbf2.flatten()
train_set_6['age_simil_35'] = rbf3.flatten()

fig, ax1 = plt.subplots()

ax1.set_xlabel("Housing median age")
ax1.set_ylabel("Number of districts")
ax1.hist(housing["housing_median_age"], bins=50)

ax2 = ax1.twinx()  # create a twin axis that shares the same x-axis
color = "red"
ax2.plot(ages, train_set_6['age_simil_16'], color=color, label="about 16")
ax2.plot(ages, train_set_6['age_simil_26'], color=color, label="about 26", linestyle="--")
ax2.plot(ages, train_set_6['age_simil_35'], color=color, label="about 35", linestyle="dotted")
ax2.tick_params(axis='y', labelcolor=color)
ax2.set_ylabel("Age similarity", color=color)

plt.legend(loc="upper left")

plt.show()






class StandardScalerClone(BaseEstimator, TransformerMixin):
    def __init__(self, with_mean=True):  # no *args or **kwargs!
        self.with_mean = with_mean

    def fit(self, X, y=None):  # y is required even though we don't use it
        X = check_array(X)  # checks that X is an array with finite float values
        self.mean_ = X.mean(axis=0)
        self.scale_ = X.std(axis=0)
        self.n_features_in_ = X.shape[1]  # every estimator stores this in fit()
        return self  # always return self!

    def transform(self, X):
        check_is_fitted(self)  # looks for learned attributes (with trailing _)
        X = check_array(X)
        assert self.n_features_in_ == X.shape[1]
        if self.with_mean:
            X = X - self.mean_
        return X / self.scale_


class ClusterSimilarity(BaseEstimator, TransformerMixin):
    def __init__(self, n_clusters=10, gamma=1.0, random_state=None):
        self.n_clusters = n_clusters
        self.gamma = gamma
        self.random_state = random_state

    def fit(self, X, y=None, sample_weight=None):
        self.kmeans_ = KMeans(self.n_clusters, n_init=10,
                              random_state=self.random_state)
        self.kmeans_.fit(X, sample_weight=sample_weight)
        return self  # always return self!

    def transform(self, X):
        return rbf_kernel(X, self.kmeans_.cluster_centers_, gamma=self.gamma)
    
    def get_feature_names_out(self, names=None):
        return [f"Cluster {i} similarity" for i in range(self.n_clusters)]

print("train_set_6_new_num_imputed_std_scaled_clone:")
std_scaler_clone = StandardScalerClone()
train_set_6_new_num_imputed_std_scaled_clone = std_scaler_clone.fit_transform(train_set_6_new_num.dropna())
cluster_simil = ClusterSimilarity(n_clusters=10, gamma=1., random_state=42)
similarities = cluster_simil.fit_transform(train_set_6[["latitude", "longitude"]],
                                           sample_weight=train_set_new_labels)

print(train_set_6_new_num_imputed_std_scaled_clone)
print("similarities[:3].round(2)")
print(similarities[:3].round(2))



def column_ratio(X):
    return X[:, [0]] / X[:, [1]]

def ratio_name(function_transformer, feature_names_in):
    return ["ratio"]  # feature names out

def ratio_pipeline():
    return make_pipeline(
        SimpleImputer(strategy="median"),
        FunctionTransformer(column_ratio, feature_names_out=ratio_name),
        StandardScaler())


log_pipeline = make_pipeline(
    SimpleImputer(strategy="median"),
    FunctionTransformer(np.log, feature_names_out="one-to-one"),
    StandardScaler())

default_num_pipeline = make_pipeline(SimpleImputer(strategy="median"),
                                     StandardScaler())

cluster_simil = ClusterSimilarity(n_clusters=10, gamma=1., random_state=42)
cat_pipeline = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OneHotEncoder(handle_unknown="ignore"))
default_num_pipeline = make_pipeline(SimpleImputer(strategy="median"),
                                     StandardScaler())
preprocessing = ColumnTransformer([
        ("bedrooms", ratio_pipeline(), ["total_bedrooms", "total_rooms"]),
        ("rooms_per_house", ratio_pipeline(), ["total_rooms", "households"]),
        ("people_per_house", ratio_pipeline(), ["population", "households"]),
        ("log", log_pipeline, ["total_bedrooms", "total_rooms", "population",
                               "households", "median_income"]),
        ("geo", cluster_simil, ["latitude", "longitude"]),
        ("cat", cat_pipeline, make_column_selector(dtype_include=object)),
    ],
    remainder=default_num_pipeline)


print("train_set_6_new_prepared_3.shape:")
train_set_6_new_prepared_3 = preprocessing.fit_transform(train_set_6)
print(train_set_6_new_prepared_3.shape)
print("preprocessing_3.get_feature_names_out():")

print(preprocessing.get_feature_names_out())



lin_reg = make_pipeline(preprocessing, LinearRegression())
lin_reg.fit(train_set_new, train_set_new_labels)
housing_predictions = lin_reg.predict(train_set_new)
print("housing_predictions[:5].round(0):")
print(housing_predictions[:5].round(0))
print("housing_labels.iloc[:5].values:")
print(train_set_new_labels.iloc[:5].values)


lin_rmse = mean_squared_error(train_set_new_labels, housing_predictions,
                              squared=False)

lin_rmses = -cross_val_score(lin_reg, train_set_new, train_set_new_labels,
                              scoring="neg_root_mean_squared_error", cv=10)

print("lin_reg_rmse:")
print(lin_rmse)
print("lin_reg_rmse_cv:")
print(lin_rmses)
print("lin_reg_rmse_cv.mean():")
print(lin_rmses.mean())


tree_reg = make_pipeline(preprocessing, DecisionTreeRegressor(random_state=42))
tree_reg.fit(train_set_new, train_set_new_labels)

predictions_tree_reg = tree_reg.predict(train_set_new)
tree_rmse = mean_squared_error(train_set_new_labels, predictions_tree_reg,
                              squared=False)
tree_rmses = -cross_val_score(tree_reg, train_set_new, train_set_new_labels,
                              scoring="neg_root_mean_squared_error", cv=10)


print("predictions_tree_reg[:5].round(0):")
print(predictions_tree_reg[:5].round(0))
print("train_set_6_new_labels[:5].values:")
print(train_set_new_labels.iloc[:5].values)
print("tree_reg_rmse:")
print(tree_rmse)
print("tree_reg_rmse_cv:")
print(tree_rmses)
print("tree_reg_rmse_cv.mean():")
print(tree_rmses.mean())



forest_reg = make_pipeline(preprocessing,
                           RandomForestRegressor(random_state=42))
forest_rmses = -cross_val_score(forest_reg, train_set_new, train_set_new_labels,
                                scoring="neg_root_mean_squared_error", cv=10)
forest_reg.fit(train_set_new, train_set_new_labels)
predictions_forest_reg = forest_reg.predict(train_set_new)
forest_rmse = mean_squared_error(train_set_new_labels, housing_predictions,
                                 squared=False)
print("predictions_forest_reg[:5].round(0):")
print(predictions_forest_reg[:5].round(0))
print("train_set_6_new_labels[:5].values:")
print(train_set_new_labels.iloc[:5].values)
print("forest_reg_rmse:")
print(forest_rmse)
print("forest_reg_rmse_cv:")
print(forest_rmses)
print("forest_reg_rmse_cv.mean():")
print(forest_rmses.mean())



full_pipeline = Pipeline([
    ("preprocessing", preprocessing),
    ("random_forest", RandomForestRegressor(random_state=42)),
])
param_grid = [
    {'preprocessing__geo__n_clusters': [5, 8, 10],
     'random_forest__max_features': [4, 6, 8]},
    {'preprocessing__geo__n_clusters': [10, 15],
     'random_forest__max_features': [6, 8, 10]},
]
grid_search = GridSearchCV(full_pipeline, param_grid, cv=3,
                           scoring='neg_root_mean_squared_error')
grid_search.fit(train_set_new, train_set_new_labels)
grid_search_results = pd.DataFrame(grid_search.cv_results_)
grid_search_results.sort_values(by="mean_test_score"
, ascending=False, inplace=True)

grid_search_results_short \
= grid_search_results[["param_preprocessing__geo__n_clusters"
, \
"param_random_forest__max_features"
, \
"split0_test_score", \
"split1_test_score", \
"split2_test_score", \
"mean_test_score"]]

print("grid_search_results_short:")
print(grid_search_results_short)
print("grid_search.best_params_:")
print(grid_search.best_params_)


final_model = grid_search.best_estimator_
predictions_final_model = final_model.predict(X=test_set_new)
final_model_rmse = mean_squared_error(y_true=test_set_new_labels, \
y_pred=predictions_final_model, \
squared=False)
print("predictions_final_model[:5].round(0):")
print(predictions_final_model[:5].round(0))
print("test_set_new_labels[:5].values:")
print(test_set_new_labels[:5].values)
print("final_model_rmse:")
print(final_model_rmse)
```
